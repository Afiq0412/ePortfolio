### Exercise 3: Tuple Unpacking
1. Define a tuple `info` with the following elements: `("John Doe", 29, "Data Scientist")`.

3. Print each variable on a new line.



```python
info = ("John Doe", 29, "Data Scientist")
name, age, profession = info

print(name)
print(age)
print(profession)
```
