### Exercise 1: Filter List with Lambda and Regular Expression
Given a list of strings, filter out only those strings that contain at least one digit.

Instructions:

1. Use a lambda function to filter the list.
```words = ["apple", "banana123", "cherry", "42mango", "grape", "kiwi9"]```\
2. Use a regular expression to check if a string contains at least one digit.


````python
import re

# Sample list
words = ["apple", "banana123", "cherry", "42mango", "grape", "kiwi9"]

# Filter words that contain at least one digit
filtered_words = list(filter(lambda x: re.search(r'\d', x), words))

print(filtered_words)  # Expected output: ['banana123', '42mango', 'kiwi9']
````

### Exercise 2: Validate Email Addresses

Write a function that validates if each email address in a list is correctly formatted using a lambda function and a regular expression.

Instructions:

1. Define a regular expression that matches standard email formats.
2. Use filter with a lambda function to select only valid emails.

#### Sample list of emails
```emails = ["test@example.com", "invalid-email@", "user@domain.co", "name@domain", "contact@website.com"]```

#### Regular expression to match valid email formats
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$' 


````python
import re

# Sample list of emails
emails = ["test@example.com", "invalid-email@", "user@domain.co", "name@domain", "contact@website.com"]

# Regular expression to match valid email formats
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

# Filter valid emails
valid_emails = list(filter(lambda email: re.match(email_pattern, email), emails))

print(valid_emails)
````

### Exercise 3: Replace pattern in string

Using lambda functions and regular expressions, write a function to replace any sequence of multiple spaces in a string with a single space.

Instructions:

1. Use ```re.sub()``` within a lambda function to replace sequences of spaces.
2. Apply it to a list of strings using map().

#### Sample list of strings with multiple spaces
```sentences = ["This   is   an example.", "Another   sentence  with   extra spaces."]```


````python
import re

# Sample list of strings with multiple spaces
sentences = ["This   is   an example.", "Another   sentence  with   extra spaces."]

# Replace multiple spaces with a single space
cleaned_sentences = list(map(lambda s: re.sub(r'\s+', ' ', s), sentences))

print(cleaned_sentences)  
# Expected output: ['This is an example.', 'Another sentence with extra spaces.']
````
