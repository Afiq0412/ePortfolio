### Exercise 1: Tuple Manipulation
1. Create a tuple `my_tuple` with the following elements: `("apple", "banana", "cherry", "date", "elderberry")`.
2. Extract the second and fourth items and store them in a new tuple.
3. Check if the item `"banana"` exists in `my_tuple`. If it does, print `"Found 'banana'!"`.
4. Convert `my_tuple` into a list, add the item `"fig"` to the end, and convert it back to a tuple.



```python
my_tuple = ("apple","banana","cherry","date","elderberry")
type(my_tuple)
print(my_tuple)

new_tuple = (my_tuple[1],  my_tuple[3])
print(type(new_tuple))
print(new_tuple)

#help(tuple)
if 'banana' in my_tuple:
    print("Found 'banana'!")
else:
    print("Not found 'banana'!")

list_convert = list(my_tuple)
list_convert.append("fig")
my_tuple = tuple(list_convert)
print(my_tuple)
```

### Exercise 2: List Slicing and Modification

5. Reverse `my_list` and print the result.




```python
my_list = [2,4,6,8,10,12,14,16]
type(my_list)
first_three = my_list[:3]
print(first_three)

print(my_list.pop())
print(my_list)

my_list.insert(1, 18)
print(my_list)

print(my_list[::-1])
```

```python
def split_list(a_list):
    mid = len(a_list)//2
    return (a_list[:mid], a_list[mid:])

split_list([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
def list_to_tuple(list1):
    unique = []
    for i in list1:
        if i not in unique:
            unique.append(i)
    return unique

list_to_tuple([1, 2, 3, 1, 2, 3, 4, 4, 5, 6, 8, 9, 1, 2])
```

```python
nums = tuple(range(1, 11))
nums

#even_squares = [x**2 for x in nums if x%2 == 0]
even_squares = list(map(lambda x: x**2, filter(lambda x: x%2 == 0, nums)))
print(even_squares)

print(tuple(even_squares))
```
