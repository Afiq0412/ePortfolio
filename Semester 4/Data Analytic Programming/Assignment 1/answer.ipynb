#### QUESTION 2: Read and parse data from the```"ride.csv"```. Complete the given code (8 marks)

````python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file) 
        #read and parse data iteratively  (5 marks)
        for row in reader:
            row['distance'] = float(row['distance'])
            row['duration'] = int(row['duration'])
            row['fare'] = round(float(row['fare']), 2)
            data.append(row)
    return data

# Write code(s) to load data from actual CSV file (2 marks)
parsed_ride_data = load_ride_data("ride.csv")

# Write code(s) to preview data (1 mark)
for ride in parsed_ride_data:
    print(ride)
````

#### QUESTION 3: Define a function to compute average fare by city and call the function by passing the data (10 marks)

````python
def average_fare_city(data):
    city_fare = {}
    city_count = {}

    for ride in data:
        city = ride['city']
        fare = ride['fare']

        if city not in city_fare:
            city_fare[city] = 0
            city_count[city] = 0

        city_fare[city] += fare
        city_count[city] += 1

    # Calculate average after processing all rides
    average_fare = {}
    for city in city_fare:
        average_fare[city] = city_fare[city] / city_count[city]
        
    return average_fare


# Write the code(s) to call your function and pass the data (2 marks)
average_fare = average_fare_city(parsed_ride_data)
print(f"Average fare by city: {average_fare}")
````

#### QUESTION 4: Define a function to summarize driver information, call the function by passing data ad display the summary (10 marks)
#### ```driver_name, num_of_rides, total_distance, total_fare```

````python

# Write your function definition here (7 marks)
def summarize_driver_information(data):
    driver_summary = {}
    num_of_rides = {}
    total_distance = {}
    total_fare = {}

    for ride in data:
        name = ride['driver_name']
        distance = ride['distance']
        fare = ride['fare']

        if name not in driver_summary:
            num_of_rides[name] = 0
            total_distance[name] = 0 
            total_fare[name] = 0
        
        num_of_rides[name] += 1
        total_distance[name] += distance
        total_fare[name] += fare
        
        driver_summary[name] = {
            'driver_name': name,
            'num_of_rides': num_of_rides[name],
            'total_distance': total_distance[name],
            'total_fare': total_fare[name]
        }
    return driver_summary

# Write the code(s) to call your function and pass the data (2 marks)
summarize_driver_information = summarize_driver_information(parsed_ride_data)
print



# Write the code(s) to display driver's summary (1 mark)
for driver in summarize_driver_information:
    print(summarize_driver_information[driver])

````

#### QUESTION 5: Implement a function ```analyze_peak_slice(data, start_idx, end_idx)``` that: (10 marks)

    1.Accepts the ride data list and two indices: start_idx and end_idx.
    2.Extracts a slice of rides from start_idx = 2 to end_idx = 6 (inclusive).
    3.Calculates and returns:
        i)  Total number of rides in that slice
        ii) Average fare
        iii)List of unique drivers involved
        
    Sample output:
 
   ```Total_rides: 5```<br>
   ```Average_fare: 18.45```<br>
   ```Drivers: Ravi, Ali, Mei``` 

````python
# Write your function definition here that accepts ride data, start index and end index, slice the list, 
# calculate and returns the values in (3)   (8 marks)
def analyze_peak_slice(data, start_idx, end_idx):

    new_data = data[start_idx:end_idx]
    total_ride = len(new_data)
    average_fare = round(((sum(x['fare'] for x in new_data)) / total_ride), 2)  
    
    drivers = []
    for ride in new_data:
        driver = ride['driver_name']
        if driver not in drivers:
            drivers.append(driver)
    
    print(f"Total rides: {total_ride}")
    print(f"Average fare: {average_fare}")
    print(f"Drivers: {', '.join(drivers)}")
    








# Write the code(s) to call your function and pass the ride data, start index and end index (2 marks)
analyze_peak_slice(parsed_ride_data, 2, 7)

````

#### QUESTION 6: Use lambda function to find top 3 drivers by total fare (7 marks)

````python
# Write your function definition here for top 3 drivers by total fare (5 marks)
def top_drivers_by_fare(data):
    sorted_drivers = sorted(data.items(), key = lambda x: x[1]['total_fare'], reverse=True)
    print(f"Sorted drivers by total fare: {sorted_drivers}")
    top_3_drivers = [driver[0] for driver in sorted_drivers[:3]]

    return top_3_drivers
    

# Write the code(s) to call your function and pass the driver's summary data (2 marks)
top_drivers = top_drivers_by_fare(summarize_driver_information)
print(f"Top 3 drivers by total fare: {', '.join(top_drivers)}")

````

#### QUESTION 7: Write a function to calculate average duration by cities and use lambda function to sort cities by average duration (10 marks)

````python
# Write your function definition here for average duration and sort cities (8 marks)
def average_duration_by_city(data):
    average_duration_by_city = {}
    city_duration = {}
    city_count = {}


    for ride in data:
        city = ride['city']
        duration = ride['duration']
    
        if city not in city_duration:
            city_duration[city] = 0
            city_count[city] = 0
            average_duration_by_city[city] = 0

        city_duration[city] += duration
        city_count[city] += 1
    
    for city in city_duration:
        average_duration_by_city[city] = city_duration[city] / city_count[city]

    sorted_cities_duration = sorted(average_duration_by_city.items(), key = lambda x: x[1], reverse=True)
    d = dict(sorted_cities_duration)


    return d



# Write the code(s) to call your function and pass the data (2 marks)
d = average_duration_by_city(parsed_ride_data)

for city, duration in d.items():
    print(city, duration)
````

#### QUESTION 8: Define a function to write driver summary to a text file ```driver_summary.txt``` with the following information:
#### ```"Driver", "Total Rides", "Total Distance", "Total Fare"```

````python
# Write your function definition here for writing driver summary into a text file (8 marks)
def driver_summary(filename, data):
    with open(filename, 'w') as driver_summary_file:
        for driver, summary in data.items():
            driver_summary_file.write(f"Driver: {driver} ")
            driver_summary_file.write(f"Total Rides: {summary['num_of_rides']} ")
            driver_summary_file.write(f"Total Distance: {summary['total_distance']} ")
            driver_summary_file.write(f"Total Fare: {summary['total_fare']}\n")


# Write the code(s) to call your function and pass the driver summary data (2 marks)
driver_summary("driver_summary.txt", summarize_driver_information)




````
