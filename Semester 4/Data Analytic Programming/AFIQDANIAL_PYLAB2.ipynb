```python
# (EXERCISE 1: LIST) #
```

```python
#1. Create and Print a List
colors = ["Red", "Blue", "Green", "Yellow", "Purple"]
print(colors)
```

```python
# 2. Accessing List Element
fruits = ["apple", "banana", "cherry", "orange", "grape"]

print(fruits[0])
print(fruits[-1])
print(fruits[::2])
```

```python
# 3. Modifying a List
numbers = [2, 4, 6, 8]

numbers.append(10)
numbers[1] = 5
numbers.pop()
print(numbers)
```

```python
# 4. Iterating Through List
names = ["Alice", "Bob", "Charlie", "David"]
for name in names:
    print(name)
```

```python
# 5. Check if an Element exist in a List
numbers = [1, 3, 5, 7, 9]

user_input = int(input("Enter a number: "))
if user_input in numbers:
    print(f"{user_input} is in the list.")
else:
    print(f"{user_input} is not in the list.")
```

```python
# (EXERCISE 2: TUPLE) #
```

```python
# 1. Create and Print a Tuple
animals = ("Dog", "Cat", "Elephant", "Tiger")
print(animals)
```

```python
# 2. Access Tuple Elements
months = ("Jan", "Feb", "Mar", "Apr")
print(months[1])
print(months[-1])
```

```python
# 3. Check Tuple Immutability
months = ("Jan", "Feb", "Mar", "Apr")
months[1] = "December"
```

```python
# 4. Tuple Unpacking
numbers = (10, 20, 30)
a, b, c = numbers
print(a, b, c)
```

```python
# 5. Convert a Tuple to a List and Modify
tup = (100, 200, 300, 400)
lst = list(tup)
lst[1] = 250
tup = tuple(lst)
print(tup)
```

```python
# (EXERCISE 3: SET) #
```

```python
# 1. Create and Print a Set
numbers = {10, 20, 30, 40, 50}
print(numbers)
```

```python
# 2. Add and Remove Elements from a Set
my_set = {1, 2, 3}
my_set.add(4)
my_set.remove(2)
print(my_set)
```

```python
# 3. Find the Union of Two Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)
```

```python
# 4. Find the Intersection of Two Sets
set1 = {10, 20, 30}
set2 = {20, 30, 40}
intersection_set = set1 & set2
print(intersection_set)
```

```python
# 5. Remove Duplicates from a List using a Set
numbers = [1, 2, 2, 3, 4, 4, 5, 5]
unique_numbers = set(numbers)
print(unique_numbers)
```
