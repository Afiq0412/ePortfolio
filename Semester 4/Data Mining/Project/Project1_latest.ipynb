```python
import numpy as np
import pandas as pd
```

```python
heart_disease = pd.read_csv("heart_disease.csv")
df = heart_disease.copy()
df
```

```python
df.head()
```

```python
checknull = df.isnull().sum()
checknull
```

```python
missing_percent = df.isnull().mean() * 100
print(missing_percent)
```

```python
# fill in the numeric columns with mean
numeric_cols = df.select_dtypes(include=['number']).columns
print(numeric_cols)
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print(df)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# fill in the categorical columns with mode

df.drop(columns=["Alcohol Consumption"], inplace=True)

categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
for col in categorical_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Define your column categories based on your data
binary_cols = [
    'Gender', 'Smoking', 'Family Heart Disease', 'Diabetes', 'High Blood Pressure', 
    'Low HDL Cholesterol', 'High LDL Cholesterol', 'Heart Disease Status'
]  # Yes/No and Male/Female columns

ordinal_cols = [
    'Exercise Habits','Stress Level', 'Sugar Consumption'
]  # Low/Medium/High columns

# Process each type
# Binary columns (Yes/No and Male/Female)
for col in binary_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Ordinal columns (Low/Medium/High)
ordinal_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
for col in ordinal_cols:
    df[col] = df[col].map(ordinal_mapping)

print("\nProcessing complete!")

df.head()
```

```python
target_column = 'Diabetes'

# Split into features and target
x = df.drop(target_column, axis=1)
y = df[target_column]

# Train/Test split|
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify = y, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

#Normalize data (IMPORTANT for KNN)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

#Train KNN model
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train_scaled, y_train)

#Predict
y_pred = knn.predict(x_test_scaled)

#Evaluate with Heatmap Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Create heatmap for confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['Pred No', 'Pred Yes'],
            yticklabels=['Actual No', 'Actual Yes'],
            cbar_kws={'label': 'Count'})

plt.title('Confusion Matrix Heatmap - KNN Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.tight_layout()
plt.show()

# Optional: Also create a DataFrame version for reference
cm_df = pd.DataFrame(cm, index=['Actual No', 'Actual Yes'], columns=['Pred No', 'Pred Yes'])
print("Confusion Matrix DataFrame:")
print(cm_df)
```

```python
from xgboost import XGBClassifier

# Train XGBoost model
model = XGBClassifier(random_state=42)
model.fit(x_train, y_train)

# Predict and evaluate
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)

# Create heatmap for confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['Pred No', 'Pred Yes'],
            yticklabels=['Actual No', 'Actual Yes'],
            cbar_kws={'label': 'Count'})

plt.title('Confusion Matrix Heatmap - XGBoost Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.tight_layout()
plt.show()

# Optional: Also create a DataFrame version for reference
cm_df = pd.DataFrame(cm, index=['Actual No', 'Actual Yes'], columns=['Pred No', 'Pred Yes'])
print("Confusion Matrix DataFrame:")
print(cm_df)
```

```python
from sklearn.naive_bayes import GaussianNB

# Step 5: Train Naive Bayes model
model = GaussianNB()
model.fit(x_train, y_train)

# Step 6: Predict and evaluate
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)

# Create heatmap for confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['Pred No', 'Pred Yes'],
            yticklabels=['Actual No', 'Actual Yes'],
            cbar_kws={'label': 'Count'})

plt.title('Confusion Matrix Heatmap - Naive Bayes Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.tight_layout()
plt.show()

# Optional: Also create a DataFrame version for reference
cm_df = pd.DataFrame(cm, index=['Actual No', 'Actual Yes'], columns=['Pred No', 'Pred Yes'])
print("Confusion Matrix DataFrame:")
print(cm_df)
```

```python
print(classification_report(y_test, y_pred, target_names=['No', 'Yes',], zero_division=0))
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

print("Accuracy :", round(accuracy*100,2))
print("Precision:", round(precision*100,2))
print("Recall   :", round(recall*100,2))
print("F1-Score :", round(f1*100,2))
```
