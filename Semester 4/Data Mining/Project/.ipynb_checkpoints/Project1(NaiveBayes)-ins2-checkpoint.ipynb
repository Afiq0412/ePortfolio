```python
import numpy as np
import pandas as pd
```

```python
heart_disease = pd.read_csv("heart_disease.csv")
df = heart_disease.copy() # make a copy to avoid the corruption in the original dataset
df
```

```python
df.head()
```

```python
checknull = df.isnull().sum()
checknull
```

```python
# fill in the numeric columns with mean
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print(df)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# fill in the categorical columns with mode

df.drop(columns=["Alcohol Consumption"], inplace=True)

target_column = 'Heart Disease Status' # target columns or laber
categorical_cols = df.select_dtypes(include=['object']).drop(columns=target_column).columns
for col in categorical_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix ,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score

# Step 2: Encode categorical variables
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Encode target if its categorical
target_le = LabelEncoder()
df[target_column] = target_le.fit_transform(df[target_column])

# Step 3: Split into features and target
x = df.drop(target_column, axis=1)
y = df[target_column]

# Step 4: Train/Test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Step 5: Train Naive Bayes model
model = GaussianNB()
model.fit(x_train, y_train)

# Step 6: Predict and evaluate
y_pred = model.predict(x_test)

cm = confusion_matrix(y_test, y_pred)

print(cm)
print()

cm_df = pd.DataFrame(cm, index=['Actual No', 'Actual Yes'], columns=['Pred No', 'Pred Yes'])

print(cm_df)
```

```python
print(classification_report(y_test, y_pred, target_names=['No', 'Yes'], zero_division=0))
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-Score :", f1)
```
